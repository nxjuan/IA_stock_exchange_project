```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime, timedelta

# Funo para obter os dados de uma semana atrs at hoje usando yfinance
def get_stock_data(symbol):
    end_date = datetime.now()
    start_date = end_date - timedelta(days=7)
    
    # Baixar os dados usando yfinance
    df = yf.download(symbol, start=start_date, end=end_date, interval="1h")
    
    return df

# Funo para plotar grfico de linha
def plot_stock_data(df):
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['Close'], label='Preo de Fechamento')
    plt.title('B3SA3 - Preo por Hora (ltima Semana)')
    plt.xlabel('Data e Hora')
    plt.ylabel('Preo de Fechamento (R$)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Smbolo para B3SA3
symbol = "B3SA3.SA"  # Smbolo correto para yfinance

# Obtenha os dados
df = get_stock_data(symbol)

# Verifique se os dados foram obtidos com sucesso antes de tentar plotar
if df is not None and not df.empty:
    # Gere o grfico
    plot_stock_data(df)
else:
    print("No foi possvel obter os dados do ativo.")
```
